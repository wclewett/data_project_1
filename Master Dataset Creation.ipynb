```python
# dependencies
import pandas as pd
import os
```

```python
# file path setup
funding_path = 'resources/funding_data'
success_path = 'resources/success_data'
years = ['2006','2007','2008','2009','2010']
```

```python
# load in funding data
funding = {}
for year in years:
    file_name = year + '_survey.xlsx'
    print(file_name)
    file_path = os.path.join(funding_path, file_name)
    print(file_path)
    try:
        funding[year] = pd.read_excel(file_path, 'Data')
    except:
        funding[year] = pd.read_excel(file_path, 'data')
```

```python
# Build Master Funding Set
for year in years:
    if int(year) == 2006:
        funding_master = pd.DataFrame(funding[year]['STNAME'])
        funding_master['STABR'] = funding[year]['STABR']
    for column in funding[year]:
        if column != 'STNAME':
            if column != 'FIPS':
                if column != 'STABR':
                    if column != 'SURVYEAR':
                        funding_master[year + '_' + column] = funding[year][column]
funding_master = funding_master.set_index('STNAME')
funding_master = funding_master.drop(['American Samoa', 'Guam', 'Northern Mariana Islands', 'Puerto Rico', 'Virgin Islands'])

formatted_st_names = []

for i in range(len(funding_master)):
    formatted_st_names.append(funding_master.index[i].lower())

funding_master.index = formatted_st_names

funding_master    
```

```python
# load in success data
success = {}
for year in years:
    file_name = year + '_survey.xlsx'
    print(file_name)
    file_path = os.path.join(success_path, file_name)
    print(file_path)
    success[year] = pd.read_excel(file_path)
```

```python
# Build Master Success Set
for year in years:
    if int(year) == 2006:
        success_master = pd.DataFrame(success[year]['STNAME'])
        success_master['STABR'] = success[year]['STABR']
    for column in success[year]:
        if column != 'STNAME':
            if column != 'CITY':
                if column != 'STABR':
                    if column != 'SURVYEAR':
                        if column != 'STREET':
                            if column != 'ZIP':
                                if column != 'ZIP4':
                                    if column != 'PHONE':
                                        if column != 'FIPST':
                                            if column != 'SEANAME':
                                                success_master[year + '_' + column] = success[year][column]
success_master = success_master.set_index('STNAME')
success_master = success_master.drop(['AMERICAN SAMOA', 'GUAM', 'NORTHERN MARIANAS', 'PUERTO RICO', 'VIRGIN ISLANDS', 'DOD OVERSEAS', 'BIE', 'DOD DOMESTIC'])

formatted_st_names = []

for i in range(len(success_master)):
    formatted_st_names.append(success_master.index[i].lower())

success_master.index = formatted_st_names
success_master
```

```python
# Save Datasets to CSVs
funding_output = 'datasets/master_funding.csv'
success_output = 'datasets/master_success.csv'

funding_master.to_csv(funding_output)
success_master.to_csv(success_output)
```
