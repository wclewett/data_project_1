```python
# dependencies
import pandas as pd
import os
```

```python
# file path setup
base_path = 'resources'
years = ['2008','2009','2010','2011','2012','2013','2014','2015','2016','2017']
```

```python
# load in data
survey = {}
for year in years:
    file_name = year + '_survey.xlsx'
    print(file_name)
    file_path = os.path.join(base_path, file_name)
    print(file_path)
    try:
        survey[year] = pd.read_excel(file_path, 'Data')
    except:
        survey[year] = pd.read_excel(file_path, 'data')
```

```python
# Build revenue dataset
rev_codes = ['R1A', 'R1B', 'R1C', 'R1D', 'R1E', 'R1F', 'R1G', 'R1H', 'R1I', 'R1J', 'R1K', 'R1L', 'R1M', 'R1N', 'STR1', 'R2',
             'R3', 'R4A', 'R4B', 'R4C', 'R4D', 'STR4', 'R5', 'TR']

revenue_df = pd.DataFrame(survey[year]['STNAME'])
for year in years:
    for code in rev_codes:
        revenue_df[year] = survey[year]['SURVYEAR']
        revenue_df[year + '_' + code] = survey[year][code]
        
revenue_df = revenue_df.set_index('STNAME')
revenue_df = revenue_df.drop(['American Samoa', 'Guam', 'Northern Mariana Islands', 'Puerto Rico', 'Virgin Islands'])
revenue_df
```
