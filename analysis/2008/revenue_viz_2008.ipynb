```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import scipy.stats as st
plt.style.use(['classic'])
```

```python
# Read in Revenue Files
file_2008 = "../../datasets/subsets/revenue/rev_2008.csv"

rev_2008 = pd.read_csv(file_2008)
rev_2008 = rev_2008.set_index(['STNAME'])
rev_2008
```

```python
(rev_2008.sum(axis=1)/1000000000).sort_values(ascending = False).head(5)
```

```python
summary_2008 = pd.DataFrame(rev_2008.sum(axis=1)/1000000000, columns=['Total Revenue ($B)'])
summary_2008
```

```python
summary_2008['Local Revenue ($B)'] = rev_2008.iloc[:,0:14].sum(axis=1)/1000000000
summary_2008
```

```python
summary_2008['State Revenue ($B)'] = rev_2008.iloc[:,15:16].sum(axis=1)/1000000000
summary_2008
```

```python
summary_2008['Federal Revenue ($B)'] = rev_2008.iloc[:,16:20].sum(axis=1)/1000000000
summary_2008
```

```python
summary_2008['Other Revenue ($B)'] = (rev_2008.iloc[:,14:15].sum(axis=1) + rev_2008.iloc[:,20:21].sum(axis=1))/1000000000
summary_2008
```

```python
summary_pct = pd.DataFrame(summary_2008['Local Revenue ($B)'] / summary_2008['Total Revenue ($B)'] *100, columns = ['Local Rev (%)'])
summary_pct['State Rev (%)'] = summary_2008['State Revenue ($B)'] / summary_2008['Total Revenue ($B)'] *100
summary_pct['Federal Rev (%)'] = summary_2008['Federal Revenue ($B)'] / summary_2008['Total Revenue ($B)'] *100
summary_pct['Other Rev (%)'] = summary_2008['Other Revenue ($B)'] / summary_2008['Total Revenue ($B)'] *100
summary_pct
```

```python
# Load in Dropout Rates
dropout_path = "../../datasets/subsets/success/success_df_2008.csv"

success_2008 = pd.read_csv(dropout_path)
success_2008
```

```python
summary_pct['Dropout Rate'] = ''

for i in range(len(success_2008)):
    summary_pct['Dropout Rate'][i] = success_2008['2008_DRP912'][i].astype(float)
    
summary_pct
```

```python
summary_pct.sort_values(by=['Dropout Rate'], ascending = False)
```

```python
summary_2008['Dropout Rate'] = ''

for i in range(len(success_2008)):
    summary_2008['Dropout Rate'][i] = success_2008['2008_DRP912'][i].astype(float)
    

summary_2008 = summary_2008[summary_2008['Dropout Rate'].notna()]
summary_2008['Dropout Rate'] = summary_2008['Dropout Rate'].astype(float)
summary_2008.info()
```

```python
# Total Revenue and Dropout Rate
plt.scatter(summary_2008['Total Revenue ($B)'], summary_2008['Dropout Rate'])
plt.xlabel('Total Revenue ($B)')
plt.xlim(left = 0, right = 80)
plt.ylabel('Dropout Rate (%)')
plt.title("Total Revenue vs Dropout Rate at the State Level")
plt.grid()
```

```python
x_values = summary_2008['Total Revenue ($B)']
y_values = summary_2008['Dropout Rate'].dropna()
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,1)) + "x + " + str(round(intercept,1))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(40,1.5),fontsize=15,color="red")
plt.xlabel('Total Revenue ($B)')
plt.ylabel('Dropout Rate (%)')
plt.title('Total Revenue vs Dropout Rate at the State Level')
plt.xlim(left = 0, right = 80)
plt.grid()
plt.show()
correlation = st.pearsonr(x_values, y_values)
print(f"The correlation between both factors is {round(correlation[0],4)}")
```
