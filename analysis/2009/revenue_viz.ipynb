```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress
plt.style.use (['classic'])
```

```python
# Read in Revenue Files
file_2009 = "../../datasets/subsets/revenue/rev_2009.csv"

rev_2009 = pd.read_csv(file_2009)
rev_2009 = rev_2009.set_index(['STNAME'])
rev_2009
```

```python
(rev_2009.sum(axis=1)/1000000000).sort_values(ascending = False).head(5)
```

```python
summary_2009= pd.DataFrame(rev_2009.sum(axis=1)/1000000000, columns=['Total Revenue ($B)'])
summary_2009
```

```python
summary_2009['Local Revenue ($B)'] = rev_2009.iloc[:,0:14].sum(axis=1)/1000000000
summary_2009
```

```python
summary_2009['State Revenue ($B)'] = rev_2009.iloc[:,15:16].sum(axis=1)/1000000000
summary_2009
```

```python
summary_2009['Federal Revenue ($B)'] = rev_2009.iloc[:,16:20].sum(axis=1)/1000000000
summary_2009
```

```python
summary_2009['Other Revenue ($B)'] = (rev_2009.iloc[:,14:15].sum(axis=1) + rev_2009.iloc[:,20:21].sum(axis=1))/1000000000
summary_2009
```

```python
summary_pct = pd.DataFrame(summary_2009['Local Revenue ($B)'] / summary_2009['Total Revenue ($B)'] *100, columns = ['Local Rev (%)'])
summary_pct['State Rev (%)'] = summary_2009['State Revenue ($B)'] / summary_2009['Total Revenue ($B)'] *100
summary_pct['Federal Rev (%)'] = summary_2009['Federal Revenue ($B)'] / summary_2009['Total Revenue ($B)'] *100
summary_pct['Other Rev (%)'] = summary_2009['Other Revenue ($B)'] / summary_2009['Total Revenue ($B)'] *100
summary_pct
```

```python
# Load in Dropout Rates
dropout_path = "../../datasets/subsets/success/success_df_2009.csv"

success_2009 = pd.read_csv(dropout_path)
success_2009
```

```python
summary_pct['Dropout Rate'] = ''

for i in range(len(success_2009)):
    summary_pct['Dropout Rate'][i] = success_2009['2009_DRP912'][i].astype(float)
    
summary_pct
```

```python
summary_pct.sort_values(by=['Dropout Rate'], ascending = False)
```

```python
summary_2009['Dropout Rate'] = ''

for i in range(len(success_2009)):
    summary_2009['Dropout Rate'][i] = success_2009['2009_DRP912'][i].astype(float)
    
summary_2009 = summary_2009[summary_2009['Dropout Rate'].notna()]
summary_2009['Dropout Rate'] = summary_2009['Dropout Rate'].astype(float)
summary_2009.info()
```

```python
# Total Revenue v Dropout

x_values = summary_2009['Total Revenue ($B)']
y_values = summary_2009['Dropout Rate'].dropna()
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(70,-55),fontsize=15,color="red")
plt.xlabel('Total Revenue ($B)')
plt.ylabel('Dropout Rate (%)')
plt.title('Total Revenue vs Dropout Rate at the State Level')
plt.grid()
plt.show()
correlation = st.pearsonr(x_values, y_values)
print(f"The correlation between both factors is {round(correlation[0],4)}")

```
